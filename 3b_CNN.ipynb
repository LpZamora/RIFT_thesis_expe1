```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_train = pd.read_csv('/home/lea/Documents/obsidian_notes/masterADS/ads_thesis_RIFT/X_train.csv')
x_test = pd.read_csv('/home/lea/Documents/obsidian_notes/masterADS/ads_thesis_RIFT/X_test.csv')
y_train = pd.read_csv('/home/lea/Documents/obsidian_notes/masterADS/ads_thesis_RIFT/y_train.csv')
y_test = pd.read_csv('/home/lea/Documents/obsidian_notes/masterADS/ads_thesis_RIFT/y_test.csv')

# # Rescale X
# x_train = x_train / 255
# x_test = x_test / 255


# Convert y to categorical
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
x_train
```

```python
model = keras.Sequential() 
model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation="relu", input_shape=(175, 240, 1))) 
model.add(keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation="relu")) 
model.add(keras.layers.MaxPool2D(pool_size=(2, 2))) 
model.add(keras.layers.Dropout(rate=0.25))
model.add(keras.layers.Flatten()) 
model.add(keras.layers.Dense(128, activation="relu")) 
model.add(keras.layers.Dropout(rate=0.5))
model.add(keras.layers.Dense(2, activation="softmax")) 
model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.Adadelta(learning_rate=1), metrics='accuracy')

model.summary()
```

```python
history = model.fit(x_train, y_train, batch_size=128, epochs=6, verbose=1, validation_split=0.2)
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss', 'validation loss'], loc='upper left')
plt.show()

#summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['accuracy', 'validation accuracy'], loc='upper left')
plt.show()
```

```python
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)

#checking whether loss & accuracy are acceptable
print(loss, accuracy)
```
